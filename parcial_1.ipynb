2. Calcule las tasas de desempleo para hombres y mujeres en Risaralda, Caldas , Tolima y Quindío. ¿Cuál es el departamento con mayor y menor tasa de desempleo de los 4?

```python
codigos_region = [17, 63, 66, 73]

df_desocupados_reg = df_desocupados[df_desocupados['DPTO'].isin(codigos_region)].copy()
df_ocupados_reg = df_ocupados[df_ocupados['DPTO'].isin(codigos_region)].copy()

df_desocupados_gen = pd.merge(
    df_desocupados_reg[df_desocupados_reg["DSI"]==1], 
    df_cgsse[['DIRECTORIO','SECUENCIA_P','ORDEN','P3271']], 
    on=['DIRECTORIO','SECUENCIA_P','ORDEN'], 
    how='left')

df_ocupados_gen = pd.merge(
    df_ocupados_reg, 
    df_cgsse[['DIRECTORIO','SECUENCIA_P','ORDEN','P3271']], 
    on=['DIRECTORIO','SECUENCIA_P','ORDEN'], 
    how='left')

desocupados_stats = df_desocupados_gen.groupby(['DPTO','P3271'])['FEX_C18'].sum().reset_index(name='desocupados')
ocupados_stats = df_ocupados_gen.groupby(['DPTO','P3271'])['FEX_C18'].sum().reset_index(name='ocupados')

pea_final = pd.merge(ocupados_stats, desocupados_stats, on=['DPTO','P3271'], how='outer').fillna(0)
pea_final['PEA'] = pea_final['ocupados'] + pea_final['desocupados']
pea_final['TD'] = (pea_final['desocupados'] / pea_final['PEA']) * 100

nombres_depto = {17: 'Caldas', 63: 'Quindío', 66: 'Risaralda', 73: 'Tolima'}
pea_final['Departamento'] = pea_final['DPTO'].map(nombres_depto)
pea_final['Sexo'] = pea_final['P3271'].map({1:'Hombres', 2:'Mujeres'})

print(pea_final[['Departamento', 'Sexo', 'TD']].sort_values(['Departamento', 'Sexo']))
```

```python
td_max = pea_final.loc[pea_final.groupby("Sexo")["TD"].idxmax()]
td_min = pea_final.loc[pea_final.groupby("Sexo")["TD"].idxmin()]

print("\nMAYOR tasa de desempleo por sexo")
print(td_max[['Departamento', 'Sexo', 'TD']])

print("\nMENOR tasa de desempleo por sexo")
print(td_min[['Departamento', 'Sexo', 'TD']])
```

El porcentaje mayor de TD para las mujeres es del Tolima con 19.20% y la menor  TD es de Caldas con 7.68%.
Para los hombres la mayor TD es Quindio con 19.20% y la menor tasa de desempleo es Caldas con 5.58%

3. . Tomando en consideración únicamente los siguientes grupos de edad:

18 a 24 24 a 30 30 a 40 40 a 50 +60

Para Risaralda, Caldas , Tolima y Quindío, cual es el grupo de edad con mayor y menor tasa de desempleo (indicar en cada departamento el grupo de de edad con mayor y menor tasa de desempleo) ?

```python
deptos_interes = [17, 63, 66, 73]

def clasificar_edad_trabajo(x):
    if 18 <= x < 24:
        return "18-24"
    elif 24 <= x < 30:
        return "24-30"
    elif 30 <= x < 40:
        return "30-40"
    elif 40 <= x < 50:
        return "40-50"
    elif x >= 60:
        return "60 en adelante"
    else:
        return "Fuera de rango"

df_cgsse["grupo_edad"] = df_cgsse["P6040"].apply(clasificar_edad_trabajo)

df_desocupados_reg = df_desocupados[df_desocupados["DPTO"].isin(deptos_interes)].copy()
des_edad = pd.merge(
    df_desocupados_reg[df_desocupados_reg["DSI"]==1],
    df_cgsse[["DIRECTORIO","SECUENCIA_P","ORDEN","grupo_edad"]],
    on=["DIRECTORIO","SECUENCIA_P","ORDEN"],
    how="left"
).groupby(["DPTO","grupo_edad"])["FEX_C18"].sum().reset_index(name="desocupados")

df_ocupados_reg = df_ocupados[df_ocupados["DPTO"].isin(deptos_interes)].copy()
ocu_edad = pd.merge(
    df_ocupados_reg,
    df_cgsse[["DIRECTORIO","SECUENCIA_P","ORDEN","grupo_edad"]],
    on=["DIRECTORIO","SECUENCIA_P","ORDEN"],
    how="left"
).groupby(["DPTO","grupo_edad"])["FEX_C18"].sum().reset_index(name="ocupados")

pea_edad_reg = pd.merge(ocu_edad, des_edad, on=["DPTO","grupo_edad"], how="outer").fillna(0)
pea_edad_reg["PEA"] = pea_edad_reg["ocupados"] + pea_edad_reg["desocupados"]
pea_edad_reg["TD"] = (pea_edad_reg["desocupados"] / pea_edad_reg["PEA"]) * 100

grupos_validos = ["18-24", "24-30", "30-40", "40-50", "60 en adelante"]
pea_final = pea_edad_reg[pea_edad_reg["grupo_edad"].isin(grupos_validos)].copy()

nombres_depto = {17: 'Caldas', 63: 'Quindío', 66: 'Risaralda', 73: 'Tolima'}
pea_final['Departamento'] = pea_final['DPTO'].map(nombres_depto)

mayores_reg = pea_final.loc[pea_final.groupby("Departamento")["TD"].idxmax()]
menores_reg = pea_final.loc[pea_final.groupby("Departamento")["TD"].idxmin()]

print("\nMayor tasa de desempleo por departamento:")
print(mayores_reg[['Departamento', 'grupo_edad', 'TD']])

print("\nMenor tasa de desempleo por departamento:")
print(menores_reg[['Departamento', 'grupo_edad', 'TD']])
```

Para los departamentso de Risaralda, Caldas y Tolima con mayor desempleo es entre 18 a 24 años, mientras que en Quindio es entre los 24 y 30 años. Dado a estos calculos realizados, la menor tasa de desempleo se registra en los mayores de 60 años en Tolima, Risarlada y Caldas y en el Quindio entre los 40 y 50 años.

4.  Entre Risaralda, Caldas , Tolima y Quindío, cuál es el departamento con menor tasa de desempleo de mujeres entre los 18 y los 25 años ?

```python
# 1. Definimos los departamentos de interés
deptos_interes = [17, 63, 66, 73]

# 2. Preparamos los datos de Desocupados (DSI == 1) incluyendo Sexo y Edad
df_desocupados_reg = pd.merge(
    df_desocupados[(df_desocupados["DSI"] == 1) & (df_desocupados["DPTO"].isin(deptos_interes))],
    df_cgsse[["DIRECTORIO", "SECUENCIA_P", "ORDEN", "P3271", "P6040"]], # Agregamos P6040 aquí
    on=["DIRECTORIO", "SECUENCIA_P", "ORDEN"],
    how="left"
)

# 3. Preparamos los datos de Ocupados incluyendo Sexo y Edad
df_ocupados_reg = pd.merge(
    df_ocupados[df_ocupados["DPTO"].isin(deptos_interes)],
    df_cgsse[["DIRECTORIO", "SECUENCIA_P", "ORDEN", "P3271", "P6040"]], # Agregamos P6040 aquí
    on=["DIRECTORIO", "SECUENCIA_P", "ORDEN"],
    how="left"
)

# 4. Filtramos para MUJERES (P3271 == 2) entre 18 y 25 años
# Para Desocupadas
desem_joven = df_desocupados_reg[
    (df_desocupados_reg["P3271"] == 2) & 
    (df_desocupados_reg["P6040"].between(18, 25))
].groupby("DPTO")["FEX_C18"].sum()

# Para Ocupadas
ocu_joven = df_ocupados_reg[
    (df_ocupados_reg["P3271"] == 2) & 
    (df_ocupados_reg["P6040"].between(18, 25))
].groupby("DPTO")["FEX_C18"].sum()

# 5. Calculamos la Tasa de Desempleo (PEA = Ocupados + Desocupados)
pea_joven = ocu_joven.add(desem_joven, fill_value=0)
resultado_final = pd.DataFrame({
    "PEA": pea_joven,
    "Desempleadas": desem_joven
}).fillna(0)

resultado_final["Tasa_desempleo(%)"] = (resultado_final["Desempleadas"] / resultado_final["PEA"]) * 100

# 6. Mapear nombres y ordenar
nombres = {17: 'CALDAS', 63: 'QUINDIO', 66: 'RISARALDA', 73: 'TOLIMA'}
resultado_final["Departamento"] = resultado_final.index.map(nombres)
resultado_final = resultado_final.sort_values(by="Tasa_desempleo(%)")

# Mostrar resultado
print("=== Tasa de Desempleo Mujeres 18-25 años (Región) ===")
print(resultado_final[['Departamento', 'Tasa_desempleo(%)']])
```

Entre los 4 departamentos, Quindio es el que presenta la menor tasa de desempleo para mujeres entre los 18 y 25 años

5.¿Cuántos hombres y mujeres hay en edad de trabajar en el quindio ? ¿Qué porcentaje de la población total representan ?

```python
df_quindio = df_cgsse[df_cgsse["DPTO"] == 63].copy()

df_quindio['es_PET'] = df_quindio['P6040'] >= 15

poblacion_total_quindio = df_quindio['FEX_C18'].sum()

pet_quindio = df_quindio[df_quindio['es_PET'] == True].groupby('P3271')['FEX_C18'].sum()

hombres_pet = pet_quindio.get(1, 0)
mujeres_pet = pet_quindio.get(2, 0)
total_pet = hombres_pet + mujeres_pet

porcentaje_hombres = (hombres_pet / poblacion_total_quindio) * 100
porcentaje_mujeres = (mujeres_pet / poblacion_total_quindio) * 100
porcentaje_pet_total = (total_pet / poblacion_total_quindio) * 100

print(f"Población Total estimada en Quindío: {int(poblacion_total_quindio)}")
print(f"Hombres  edad de trabajar: {int(hombres_pet)} ({porcentaje_hombres:.2f}% del total)")
print(f"Mujeres  edad de trabajar: {int(mujeres_pet)} ({porcentaje_mujeres:.2f}% del total)")
print(f"Total PET Quindío: {int(total_pet)} ({porcentaje_pet_total:.2f}% del total)")
```

Con los datos alcanzados en el departamento del Quindio:
Hombres en edad de trabajar: Representan aproximadamente el 40.96%
Mujeres en edad de trabajar: Representan aproximadamente el 43.34%
Poblacion total PET: Cerca del 80-82% para el Quindio, aca los quindianos tienen mas de 15 años.

6. ¿Qué porcentaje de la población del Tolima está en edad de trabajar ?

```python
# 1. Filtrar datos para Tolima (Código 73)
df_tolima = df_cgsse[df_cgsse["DPTO"] == 73].copy()

# 2. Identificar la PET (15 años o más)
df_tolima['es_PET'] = df_tolima['P6040'] >= 15

# 3. Sumar poblaciones usando el factor de expansión (FEX_C18)
poblacion_total_tolima = df_tolima['FEX_C18'].sum()
poblacion_pet_tolima = df_tolima[df_tolima['es_PET'] == True]['FEX_C18'].sum()

# 4. Calcular el porcentaje
porcentaje_pet = (poblacion_pet_tolima / poblacion_total_tolima) * 100

print(f"Población Total Tolima: {int(poblacion_total_tolima)}")
print(f"Población en Edad de Trabajar (PET): {int(poblacion_pet_tolima)}")
print(f"Porcentaje PET: {porcentaje_pet:.2f}%")
```

En el departamento del Tolima la poblacion en edad de trabajar es del 77.01% de su poblacion
